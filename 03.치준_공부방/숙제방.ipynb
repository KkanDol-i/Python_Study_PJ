# 입력값의 마지막 숫자값을 이용한 홀짝 판별

```python
number = input("숫자를 입력해주세요")
last_character = number[-1]
last_character = int(last_character)
if last_character == 0 \
    or last_character == 2 \
    or last_character == 4 \
    or last_character == 6 \
    or last_character == 8:
    print("짝수입니다.")
else:
    print("홀수입니다.")
```

# in 문자열 연산자를 활용한 홀짝 판별
