**for 문과 range()**
range(c) : 0부터 c-1까지
인자가 하나일 때는 0부터 시작하여 입력한 숫자 직전까지의 범위를 생성합니다.

```python
c = 5
print(f"range({c}) 결과:")
for i in range(c):
    print(i, end=" ")  # 출력: 0 1 2 3 4
print("\n")
```

**for 문과 range()**
range(c, d) : c부터 d-1까지
인자가 두 개일 때는 **시작값(c)**과 **종료값(d)**을 지정합니다. 역시 종료값인 d는 포함되지 않습니다.

```python
c, d = 3, 8
print(f"range({c}, {d}) 결과:")
for i in range(c, d):
    print(i, end=" ")  # 출력: 3 4 5 6 7
print("\n")
```

**for 문과 range() **굵은 텍스트**
 ** range(c, d, e) : c부터 d-1까지 e만큼씩 건너뛰기
 인자가 세 개일 때는 마지막 인자 **e가 증감폭(step)**이 됩니다. 양수면 증가, 음수면 감소합니다.

증가하는 경우 (e > 0):

```python
c, d, e = 1, 10, 2
print(f"range({c}, {d}, {e}) 결과 (증가):")
for i in range(c, d, e):
    print(i, end=" ")  # 출력: 1 3 5 7 9
print("\n")
```

감소하는 경우 (e < 0):

```python
c, d, e = 10, 0, -2
print(f"range({c}, {d}, {e}) 결과 (감소):")
for i in range(c, d, e):
    print(i, end=" ")  # 출력: 10 8 6 4 2
print("\n")
```

예제 4-4: : 두 정수를 입력 받아 최대 공약수를 구하는 프로그램을 작성<br>
알고리즘<br>
① 최대공약수(gcd)와 반복카운터변수(step) 초기화 한 후, 두 정수를 입력 받는다. <br>
② step이 두 정수보다 작은지를 비교한다.<br>
③ ②조건을 만족할 경우 두 정수가 step으로 나눠지면 최대공약수를 변경한다.<br>
④ ②조건을 만족하지 않을 때까지 ③을  반복한다.<br>
⑤ ②조건을 만족하지 않으면 현재 최대공약수를 출력하고 종료한다.<br>

```python
# 1. 초기값 설정
gcd = 1
step = 2

# 2. 사용자로부터 두 숫자 입력 받기
# 입력창이 뜨면 숫자를 치고 '엔터'를 눌러주세요.
num1 = int(input('첫 번째 숫자를 입력하세요: '))
num2 = int(input('두 번째 숫자를 입력하세요: '))

# 3. 최대공약수 찾기 (반복문)
while (step <= num1) and (step <= num2):
    # 만약 step이 두 수의 공약수라면
    if (num1 % step == 0) and (num2 % step == 0):
        gcd = step

    # [수정된 부분] if문 밖으로 꺼내서 무조건 1씩 증가하게 함
    # 이 줄이 안쪽에 있으면 코드가 끝나지 않고 계속 돌아갑니다!
    step = step + 1

# 4. 결과 출력
print('최대공약수 : ', gcd)
```

```python
#4-2 : 입려한 수 중 가장 큰값 출력 하기
print("5개 정수를 입력하면 가장 큰 값을 화면에 출력합니다")
maxValue = 0
count = 1
while count < 6:
    n = int(input("0보다 큰 정수 한 개를 입력하세요: "))
    if n > maxValue:
        maxValue = n
    count += 1
print("최대값 = %d" %maxValue)
```

```python
#4-5 숫자피라미드 출력하기
# 숫자 피라미드 출력
n = int(input("피라미드의 높이를 입력하세요: "))

for i in range(1, n + 1):  # 1부터 n까지 반복
    for j in range(1, i + 1):  # 각 줄에서 숫자를 1부터 i까지 출력
        print(j, end=" ")
    print()  # 줄바꿈
```
