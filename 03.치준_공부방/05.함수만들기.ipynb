```python
# 가변 매개변수
def print_n_times(n, *values):
    for i in range(n):
        for j in values:
            print(j)
        print()
print_n_times(1, "안녕하세요", "즐거운", "파이썬 프로그래밍","잘가")
```

```python
#  기본 매개변수
def print_n_times(n=2, *values):
    for i in range(n):
        for j in values:
            print(j)
        print()
print_n_times( "안녕하세요", "즐거운", "파이썬 프로그래밍","잘가")
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n*factorial(n-1)
factorial(4)
```

```python
# 문제: 상대방이 원하는 팩토리얼을 구해주기
# 프로그램 코딩 과정

# 1-1입력 받기
inputdata = input("당신이 원하는 팩토리얼을 숫자로 입력해주세요.")

# 1-2문자 -> 숫자

inputdata = int(inputdata)


# 2.팩토리얼 구하기

def factorial_test(n):
    outcome = 1
    for i in range( n ,0,-1):
        outcome *=i  
    return outcome





# 3.받은 숫자를 파라미터 값으로 설정하기 
output = factorial_test(inputdata)


# 4.구한 팩토리얼 출력해주기
print("당신이 선택한 {}! 결과값은 {}입니다".format(inputdata,output))




```

```python
def power(item):
    return item * item
def under_3(item):
    return item < 3
list_input_a = [1,2,3,4,5]
output_a = map(power,list_input_a)
print(list(output_a))
output_b = filter(under_3,list_input_a)
print(list(output_b))
```
