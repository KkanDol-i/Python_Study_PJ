````python
print("== Program Start")

a = 99/0            # error
print(f"99/0 : {a}")

print("== Program End")
````

1-2) 에러가 발생했을 때 프로그램이 죽지 않도록 한다! 그것이 예외처리

````python
print("== Program Start")

try:
    a = 99/0            # error
    print(f"99/0 : {a}")
except:
    print("== error!!  but, still alive")

print("== Program End")
````

2. 파이썬 예외 처리 try, except, else, finally
2-1) 파이썬 try, except 예외처리

````python
print("== Program Start")

try:
    arr = ['b', 'l', 'o', 'g']
    print(arr[8]) # error
    print("== Mid")
except:
    print("== error!!  but, still alive")

print("== Program End")
````

2-2) 파이썬 예외처리 try, except, else

````python
print("== Program Start")

try:
    arr = ['b', 'l', 'o', 'g']
    print(arr[0])
    print("== Mid")
except:
    print("== error!!  but, still alive")
else:
    print("== else")

print("== Program End")
````

2-3) try + finally 구문

````python
print("== Program Start")

try:
    arr = ['b', 'l', 'o', 'g']
    print(arr[8])  # error
    print("== Mid")
finally:
    print("== finally")

print("== Program End")
````

2-4) try + except + finally

````python
print("== Program Start")

try:
    arr = ['b', 'l', 'o', 'g']
    print(arr[8])  # error
    print("== Mid")
except:
    print("== except")
finally:
    print("== finally")

print("== Program End")
````

2-5) try + except + else + finally

````python
print("== Program Start")

try:
    arr = ['b', 'l', 'o', 'g']
    # print(arr[0]) # not error
    # print(arr[8]) # error
    print("== Mid")
except:
    print("== except")
else:
    print("== else")
finally:
    print("== finally")

print("== Program End")
````
