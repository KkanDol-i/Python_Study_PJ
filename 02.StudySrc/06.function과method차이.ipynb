함수 : function 은 독립체 
메서드 : (Method) 객체 안에 있는 함수를 메서드라 부른다.
생성자 : 객체가 호출 될때 무조건 실행해 주는 객체 내부의 함수이다.
인스턴스 메서드 : 객체 내부에 존재하는 함수를 지칭 한다.
클래스 메서드 : 파라미터 인자에 객체를 넘기며, 해당 객체를 생성하여, 생성자를 설정 이후에 생성된 객체의 내부 인스턴스를 사용할수 있다. (재귀형)
정적 메서드 : 

개념  
 - 객체(Class) > 생성자 > 인스턴스 메서드

```python
# 함수 (Function)
def calculate_sum(a, b):
    return a + b

result = calculate_sum(5, 3)  # 8


# 메서드 (Method)
class Calculator:
    def __init__(self):
        self.value = 0
        print("Calculator initialized  : " + str(self.value))

    # 인스턴스 메서드
    def add(self, num):
        self.value += num
        print("add called: " + str(self.value))
        return self.value

    # 클래스 메서드
    @classmethod
    def from_value(cls, initial_value):
        calc = cls()
        calc.value = initial_value
        print("from_value called: " + str(calc.value))
        return calc

    # 정적 메서드 (self 없음)
    @staticmethod
    def is_positive(num):
        print("is_positive called: " + str(num > 0))
        return num > 0

# 사용 예시
calc = Calculator()
calc.add(5)  # 인스턴스 메서드 호출
Calculator.from_value(10)  # 클래스 메서드 호출
Calculator.is_positive(3)  # 정적 메서드 호출
```
