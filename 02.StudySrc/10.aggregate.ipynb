```python
import pandas as pd
import numpy as np

df = pd.DataFrame([[1,4,7],[2,5,8],[3,6,9]])
print(df)
```

```python
ex1 = df.agg(np.prod)
print(ex1)
```

```python
ex2 = df.agg('prod')
print(ex2)
```

```python
ex3 = df.agg([lambda x : min(x) * max(x)])
print(ex3)
```

```python
# list 형태의 집계 함수 전달

ex6 = df.agg(['min','max','sum','prod'])
print(ex6)
```

```python
# dict 형태의 집계 함수 전달
ex7 = df.agg({2:'sum',0:'max',1:'min'})
print(ex7)
```
