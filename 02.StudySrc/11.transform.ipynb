```python
import pandas as pd
import numpy as np

col = ['col1','col2','col3']
row = ['row1','row2','row3']
df = pd.DataFrame(data=[[10,40,70],[20,50,80],[30,60,90]],index=row,columns=col)
print(df)
```

```python
# np.sqrt 함수를 DataFrame에 각 원소별로 적용
ex1 = df.transform(np.sqrt)
print(ex1)
```

```python
# 'sqrt' 문자열을 사용하여 제곱근 함수 적용
ex2 = df.transform('sqrt')
print(ex2)
```
